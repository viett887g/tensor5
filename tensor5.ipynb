```python
from PIL import Image

# If you want a greyscale image, simply convert it to the L (Luminance) mode:
joker_a = joker.convert('L')
display(joker_a)
```

```python
#cắt ảnh 
jokermin = (1,1,150,100) 
jokercut = joker.crop(jokermin)
jokercut
```

```python
jokercut.save("anh.jpg")
```

```python
# Thu nhỏ ảnh
minimize=(50,100)
joker.thumbnail(minimize)
display(joker)
```

```python
# Phóng to ảnh
jokerbig=joker.resize((500, 500))
display(jokerbig)
```

```python
# Lật ảnh:
display(joker.transpose(Image.FLIP_TOP_BOTTOM))
```
